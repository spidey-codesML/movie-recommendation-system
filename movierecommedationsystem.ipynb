```python
#Hello geek I am Navin prasath , this code was done by me and my friend Nithesh B, a Movie recommendation system that returns all the similar movies for the input movie name you give.
#Importing the required libraries
import numpy as np #used to create numpy arrays
import pandas as pd #used to create dataframes i.e structured tables
import difflib #used to find the close name of the movie name given by the user in the dataset
from sklearn.feature_extraction.text import TfidfVectorizer #converting textual data to feature vectors
from sklearn.metrics.pairwise import cosine_similarity #used to find the similarity confidence score to recommend similar movies
```

```python
#Data Collection and Pre-processing
movies_data = pd.read_csv('/content/movies.csv')#loading the data from the csv file to a pandas dataframe
```

```python
movies_data.head(10) #printing the first 10 rows of the dataframe
```

```python
movies_data.shape #displays the number of rows and columns in the dataframe
selected_features = ['genres','keywords','tagline','cast','director'] #selecting the relevant features for recommendation(feature selection)
print(selected_features)
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')#replacing the null values with null strings , fillna() means filling the not available values
```

```python
#combining all the 5 selected features(columns)
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
combined_features = combined_features.fillna('')
print(combined_features)
```

```python
#we are converting the text data to feature vector to find the cosine similarity value
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features) #variable to store all the vectorized numerical values
print(feature_vectors)
```

```python
#Cosine Similarity
#getting similarity score or similarity confidence value using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
print(similarity.shape)
```

```python
#getting movie name as input from the user
movie_name = input('enter your favourite movie name:')
```

```python
#creating a list with all the movie names given in the dataset
all_movie_titles = movies_data['title'].tolist()
```

```python
#finding the close match for the movie name given by the user using difflib
find_close_match = difflib.get_close_matches(movie_name,all_movie_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
#sorting the similarity score list
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
print("some related movies that you can watch: \n")
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i,'.',title_from_index)
    i+=1
```

```python
#MOVIE RECOMMENDATION SYSTEM
movie_name = input('enter your favourite movie name: ')

find_close_match = difflib.get_close_matches(movie_name,all_movie_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print("\nsome related movies that you can watch: \n")
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<31):
    print(i,'.',title_from_index)
    i+=1
```
